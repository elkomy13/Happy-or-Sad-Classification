```python
import tensorflow as tf
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imghdr


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.metrics import Precision , Recall , BinaryAccuracy
```

# Avoid OOM errors by setting the GPU Memory consumption Growth

```python
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
## Removing dodgy images

data='C:/Users/youss/OneDrive/Desktop/DL Projects/Imaga Classifier for any Image/data'
image_extensions = ['jpg', 'jpeg', 'png', 'bmp', 'gif']
for image_class in os.listdir(data):
    for image in os.listdir(os.path.join(data, image_class)):
        image_path = os.path.join(data, image_class, image)
        try:
            img=cv2.imread(image_path)
            tipo=imghdr.what(image_path)
            if tipo not in image_extensions:
                print(f"Removing {image_path}")
                os.remove(image_path)
        except Exception as e:
            print("issue with image format:".format(image_path))
```

```python
# Load the data
data=tf.keras.preprocessing.image_dataset_from_directory(data)
```

```python
# loop through the data to see the images and labels
data_iterator=data.as_numpy_iterator()
```

```python
# Image represented as numpy array
batch=data_iterator.next()

# batch contain 32 images and labels
# if we run it again we will get the next 32 images and labels
```

# Pre-processing Data

```python
# Scale the data
data=data.map(lambda x, y: (x/255, y))
```

```python
data_iterator=data.as_numpy_iterator()
batch=data_iterator.next()
```

```python
batch[0].min(), batch[0].max()
```

```python
# split the data
train_size=int(len(data)*0.7)
val_size=int(len(data)*0.15)
test_size=int(len(data)*0.15)
```

```python
train=data.take(train_size)
val=data.skip(train_size).take(val_size)
test=data.skip(train_size).skip(val_size).take(test_size)
```

## Deep Learning Model

```python
model=Sequential(
    [ # 1 --> number of stride
        Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)),
        MaxPooling2D(),

        Conv2D(32, (3,3), 1, activation='relu'),
        MaxPooling2D(),

        Conv2D(16, (3,3), 1, activation='relu'),
        MaxPooling2D(),

        Flatten(),

        Dense(256, activation='relu'),
        Dense(1,activation='sigmoid')
    ]
)
```

```python
model.compile(optimizer='adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

# Train

```python
logdir='logs'
```

```python
tensorboard_callback=tf.keras.callbacks.TensorBoard(logdir)
```

```python
history=model.fit(train, validation_data=val, epochs=20,callbacks=[tensorboard_callback])
```

# Plot perfomance

```python
fig=plt.figure()
plt.plot(history.history['loss'], label='loss', color='red')
plt.plot(history.history['val_loss'], label='val_loss', color='blue')
fig.suptitle('Loss')
plt.legend()
plt.show()
```

# Evaluate Performance

```python
pre=Precision()
rec=Recall()
acc=BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    y_pred=model.predict(X)
    pre.update_state(y, y_pred)
    rec.update_state(y, y_pred)
    acc.update_state(y, y_pred)
```

```python
print(f"Precision: {pre.result().numpy()}")
print(f"Recall: {rec.result().numpy()}")
print(f"Accuracy: {acc.result().numpy()}")
```
